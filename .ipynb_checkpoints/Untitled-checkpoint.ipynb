```python
import pandas as pd
import numpy as np
import sklearn
## For plotting
import matplotlib.pyplot as plt
import seaborn as sns
## This sets the plot style
## to have a grid on a white background
sns.set_style("whitegrid")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
```

```python
#data_corpus=["guru99 is the best sitefor online tutorials. I love to visit guru99."]

file = open('lorem_Ipsum.txt','rt')
latin = file.read()
```

```python
#Converts a text file into a list of strings containing its word
latin_list = latin.split()

vocabulary=vectorizer.fit(latin_list)
X= vectorizer.transform(latin_list)
Y = X.toarray()
print(Y.shape)
Y.max()
V = vocabulary.get_feature_names()
print(V)
```

```python
Freq = np.zeros(len(V))
temp = 0
for v in range(len(V) ):
    for j in range( len(latin_list) ):
        if(Y[j,v] != 0):
           # print(j,v)
            temp += 1
    Freq[v] = temp
    temp = 0    
plt.figure(figsize=(10,10))

#x = [0, 50, 100, 150, 200]
#labels = V
#plt.xticks(x, labels, rotation='vertical')
#plt.subplots_adjust(bottom=0.15)
#plt.xlabel("Vocabulary Items", fontsize=16)
#plt.ylabel("Frequency", fontsize=16)
plt.hist(Freq)
plt.show()
```

```python
fig, axes = plt.subplots(5, 1, figsize = (12,30))
ax = axes.ravel()
for i in range(5):

    ax[i].hist(Freq, bins = 20 + 4*i, color = 'red', alpha = .5)
    ax[i].set_yticks(())

ax[0].legend(['malignant', 'benign'], loc = 'best')
    
fig.tight_layout()
```

```python
#bar graph: frequency of each vocabulary item (i.e. word)
plt.figure(figsize=(200,200))
x = V
y = Freq
labels = V
plt.bar(x, y, align='center')
plt.xticks(x, labels, rotation='vertical', size=20)
plt.yticks(y, size=20)
plt.subplots_adjust(bottom=0.15)
plt.xlabel("Vocabulary Items", fontsize=50)
plt.ylabel("Frequency", fontsize=50)

plt.show()
```
